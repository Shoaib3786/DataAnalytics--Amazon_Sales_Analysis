Finding came up with:
1. SALES AMOUNT = (SALES PRICE) * (SALES QUANTITY)
    -- SALES AMOUNT is nothing but the Total Revenue generated from Sale price which is (< or >) than LIST PRICE.


2. SALES MARGIN AMOUNT = (SALES PRICE) - (SALES COST AMOUNT)
    -- SALES MARGIN AMOUNT is nothing the Gross Profit = total SP - product producing price.


3. DISCOUNT AMOUNT = (SALES AMOUNT BASED ON LIST PRICE) - (SALES AMOUNT)


4. SALES AMOUNT BASED ON LIST PRICE = (LIST PRICE) * (SALES QUANTITY)

SALES PRICE => Price that Amazon/ seller's decides to sell the product.

LIST PRICE => Price the is suggested by the Manufactorer to the retailer.
